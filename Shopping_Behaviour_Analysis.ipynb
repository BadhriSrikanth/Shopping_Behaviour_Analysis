# **Variable Identification**

**Dependent Variable (DV)**:
The dependent variable is the outcome or response that the study seeks to predict or explain. It changes based on the influence of independent or confounding variables.

-> ***Item Purchased**  (specific product bought)*


**Independent Variable (IV)**:
The independent variable is the factor that is manipulated or observed to determine its effect on the dependent variable. It serves as the main predictor in the analysis.

->***Season (Winter, Spring, Summer, Fall)***

**Confounding Variables:**
A confounding variable is an external factor that is not the primary focus of the study but can influence both the dependent and independent variables, potentially biasing results.

->Age (numerical, continuous)

->Gender (categorical, nominal)

->Location (categorical, nominal – different states/regions)

->Discount Applied (binary categorical: Yes/No)

->Promo Code Used (binary categorical: Yes/No)

->Subscription Status (binary categorical: Yes/No)

->Payment Method (categorical: Credit Card, PayPal, etc.)

->Color (categorical, nominal – apparel color preferences)

->Previous Purchases (numerical, discrete – shopping history/loyalty)

**Handling Confounders:**

In this study, confounding variables will be included in the analysis to control for their effects and ensure that the impact of season on clothing purchases is measured more accurately. Since their influence is difficult to predict in isolation, incorporating them in multivariate analysis reduces bias and improves model reliability.

```python
#outlier detection-Z score
import pandas as pd
import numpy as np
from scipy import stats
numeric_cols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
z_scores = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
outliers = (z_scores > 3).any(axis=1)
print(f"Number of outliers detected: {outliers.sum()}")
print("\nOutlier rows:")
print(df[outliers])


```

The outlier detection identified two records in the dataset with unusually extreme or invalid values. # PLAN FOR DATA CLEANING

To ensure the dataset is clean, consistent, and ready for analysis, a structured data cleaning plan will be implemented based on the issues identified earlier. First, missing values in columns such as Age, Item Purchased, Review Rating, and Previous Purchases will be handled appropriately — numerical missing values will be replaced using measures like the mean or median, while categorical missing values will be filled with the mode or labeled as “Unknown” when suitable.

Next, data type corrections will be performed to ensure that columns like Purchase Amount (USD) and Previous Purchases are stored as numeric types, allowing for accurate calculations and statistical analysis. Inconsistent entries, such as 'Fifty' or negative amounts like -120, will be identified and either corrected or removed.

Categorical inconsistencies will also be standardized — for example, ensuring that values like 'Y' and 'N' are replaced with 'Yes' and 'No', and correcting misspellings such as 'PayPall' to 'PayPal'. Similarly, invalid entries like '?' in the Location column will be treated as missing data.

To handle outliers, statistical methods such as the Z-score technique will be used to detect extreme values in numeric fields (e.g., Age or Purchase Amount (USD)). Outliers will then be reviewed to determine whether they represent genuine but rare data or errors requiring removal.

Finally, the dataset will be checked for duplicates, unnecessary whitespace, and uniform formatting (e.g., consistent capitalization in categorical columns). By following these steps, the data will be standardized, accurate, and suitable for reliable analysis and modeling.

```python
#data cleaning implementation
import pandas as pd
import numpy as np
from scipy import stats
# Step 1: Handle Missing Values (assign back to columns)
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Item Purchased'].fillna(df['Item Purchased'].mode()[0],inplace=True)
df['Review Rating'].fillna(df['Review Rating'].median(), inplace=True)
df['Previous Purchases'].fillna(df['Previous Purchases'].median(), inplace=True)
# Step 2: Fix Data Type Issues
df['Purchase Amount (USD)'] = pd.to_numeric(df['Purchase Amount (USD)'], errors='coerce')
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df['Purchase Amount (USD)'].median())
# Step 3: Correct Categorical Inconsistencies
df['Discount Applied'] = df['Discount Applied'].replace({'Y': 'Yes', 'N': 'No'})
df['Promo Code Used'] = df['Promo Code Used'].replace({'Y': 'Yes', 'N': 'No'})
df['Payment Method'] = df['Payment Method'].replace({'PayPall': 'PayPal', 'Debit': 'Debit Card'})
df['Frequency of Purchases'] = df['Frequency of Purchases'].replace({'Quarterl': 'Quarterly'})
df['Location'] = df['Location'].replace({'?': np.nan})
# Step 4: Handle Outliers using Z-Score
numeric_cols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
z_scores = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
df = df[(z_scores < 3).all(axis=1)]  # keep only rows within Z < 3
# Step 5: Remove Duplicates
df = df.drop_duplicates()
```

```python
#PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
import pandas as pd

df = pd.read_csv("/content/ccd.csv")

# encode all categorical columns
le = LabelEncoder()
df_encoded = df.apply(lambda col: le.fit_transform(col) if col.dtype=='object' else col)

pca = PCA()
pca.fit(df_encoded)

explained = pca.explained_variance_ratio_.cumsum()
explained
```

EXPLANATION OF PCA RESULT:

The PCA results show that the cumulative explained variance increases extremely rapidly, reaching over 99.9% with the first component and achieving 100% before all 18 components are used. At first glance, this might suggest that PCA is highly efficient at capturing variance. However, this pattern actually indicates that PCA is not providing any **meaningful dimensionality reduction for this dataset**. The primary reason is that the dataset is dominated by categorical variables that were converted into numerical values through label encoding.

 PCA assumes continuous, linear relationships, but encoded categorical values introduce artificial variance because the numerical labels do not represent true quantitative distances. As a result, the PCA algorithm interprets these jumps in label values as major sources of variation, causing the first few principal components to capture almost all variance—not because the dataset is truly low-dimensional, but because PCA is reacting to arbitrary encoding rather than meaningful structure.

Ultimately, this confirms that **PCA is not an appropriate method for dimensionality reduction in this scenario, and the dimensionality of the dataset effectively remains unchanged despite the high explained variance values**.

# 7. Running ML Code

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

df=pd.read_csv("/content/ccd.csv")
features = [
    'Category', 'Season_Score','Gender',
    'Age','Purchase Amount (USD)','Item_Score','Location','Size','Color','Frequency of Purchases', 'Discount Applied',
    'Subscription Status', ]

X=df[features]
y=df['Item Purchased']
X = X.copy()
label_encoders={}

for col in X.columns:
    if X[col].dtype =='object':
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
        label_encoders[col] = le

# Encode target
le_y = LabelEncoder()
y = le_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
dt = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=4,              # allow learning more patterns
    min_samples_split=3,
    random_state=42
)

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 12))
tree.plot_tree(dt, filled=True, feature_names=X.columns, class_names=True)
plt.show()
```
